# **Q2. t-SNE**

**Importing necsessary libraries**

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

**Importing Digits dataset**

```python
digits = load_digits()
X = digits.data
y = digits.target
print(digits.data.shape)
print(X.shape)
print(y.shape)
```

## **TSNE with 1000 iterations**

```python
tsne1000 = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
X1000 = tsne1000.fit_transform(X)
```

```python
print(X1000.view())
```

```python
df1 = pd.DataFrame(data=X1000,columns=['A', 'B'])
df1.head()
```

```python
sns.scatterplot(data=df1,x=df1.A,y=df1.B,hue=y,palette='bright',)
```

## **TSNE with 250 iterations (minimum possible iterations)**

```python
tsne250 = TSNE(n_components=2, perplexity=30, n_iter=250, random_state=42)
X250 = tsne250.fit_transform(X)
```

```python
print(X250.view())
```

```python
df2 = pd.DataFrame(data=X250,columns=['A', 'B'])
df2.head()
```

```python
sns.scatterplot(data=df2,x=df2.A,y=df2.B,hue=y,palette='bright')
```

CHANGE IN SCATTERPLOTS WITH NO. OF ITERATIONS = 250, 260, 270, 280, 290, 300

```python
for i in range(250, 310, 10):
  tsne = TSNE(n_components=2, perplexity=30, n_iter=i, random_state=42, learning_rate='auto', init='random').fit_transform(X)
  #X_new = tsne.fit_transform(X)
  df = pd.DataFrame(data=tsne,columns=['A', 'B'])
  print(f"-------------------{i} iterations--------------------")
  ax = sns.scatterplot(data=df,x=df.A,y=df.B,hue=y,palette='bright')
  plt.show()
```
